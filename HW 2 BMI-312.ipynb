```python
unique_patients = somatic.groupby('patient').count() # Returns & displays unique patients 
unique_patients
list(unique_patients[unique_patients['gene'] > 3].index) # Find patients who have more than three somatic mutations. Return the IDs as a set.
```

```python
silence = somatic[somatic['variant_class'] == 'Silent'] # Separates patients who have had a silent mutation into a list
silent = set(silence['patient']) # Turns unwanted list to a set
patient = set(somatic['patient']) # Creates a new set that we want
patient.difference(silent) # Find patients who have no silent mutations. (Hint, use a set).
```

```python
merged = clinical.join(somatic.set_index('patient'), on='patient', how='outer') # Combines the two dataframes by their shared patients
merged # Perform an outer join of “clinical” and “somatic” data on patient IDs.
```

```python
merged.shape # Number of rows of columns.
```

```python
final = merged[['patient', 'diagnosis', 'gene']].groupby(['diagnosis', 'patient']).count() 
final 
```

```python
final.loc['Breast Invasive Carcinoma'].mean() # Average number of somatic mutations for patients with cancers.
final.loc['Normal'].mean() # Average number of somatic mutations for patients without cancers.
```
